```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
raw = pd.read_csv("../output/end_to_end_measurement.csv", sep="\t", header=None)

latency_in_ms = raw[3] - raw[0]
# Plotting the time series
plt.figure(figsize=(15, 6))
plt.plot(raw.index, latency_in_ms, linestyle='-')

# Adding labels and title
plt.xlabel('Sequence of Measurements')
plt.ylabel('Latency (milliseconds)')
plt.title('End-to-End Latency Time Series')
plt.savefig('latency_measurement.png')
```

```python
latency_fragments = pd.DataFrame({"client_to_server": raw[1] - raw[0],
                                  "within_server": raw[2] - raw[1],
                                  "server_to_client": raw[3] - raw[2]
                                })

plt.figure(figsize=(15, 6))
# Plotting the three columns
plt.plot(latency_fragments["client_to_server"], label="Client to Server")
plt.plot(latency_fragments["within_server"], label="Within Server")
plt.plot(latency_fragments["server_to_client"], label="Server to Client")


# Adding labels and legend
plt.xlabel("Sequence of measurement")
plt.ylabel("Latency (milliseconds)")
plt.legend()

# Show the plot
plt.savefig("latency_divided.png")
```
